```python
import pandas as pd
import numpy as np
```

## DataFrameの作成

```python
pd.DataFrame(
    {
        '名前':   [  '佐藤',   '斎藤',   '鈴木'],
        '年齢':   [      21,       30,       18],
        '住所':   ['東京都', '岐阜県', '埼玉県'],
        '血液型': [     'A',   np.nan,      'O'],
    }
)

# indexを指定する場合
pd.DataFrame(
    {
        '名前':   [  '佐藤',   '斎藤',   '鈴木'],
        '年齢':   [      21,       30,       18],
        '住所':   ['東京都', '岐阜県', '埼玉県'],
        '血液型': [     'A',   np.nan,      'O'],
    },index=      [   'i-1',    'i-2',    'i-3']
)
```

## Seriesの作成

```python
pd.Series(
    ['佐藤', '斎藤', '鈴木']
)

# indexを指定する場合
pd.Series(
    ['佐藤', '斎藤', '鈴木'],
    index=['i-1', 'i-2', 'i-3']
)
```

## ファイル読み込み
以下のファイルを読み込むことができる
- CSV
- Excel
- HTML
- json
- xml
- pickle
など

```python
# 読み込み時にindex扱いする列を指定することも可
poke_DF = pd.read_csv('../dataset/pokemon_data.csv', index_col='ぜんこくNo.')
display(poke_DF)
```

## DataFrameから行・列を抽出

```python
display(poke_DF.loc[:, ['名前']])
display(poke_DF.loc[[1], :])
display(poke_DF.loc[1 : 3, '名前' : '英語名'])
```

```python
# Series型で抽出する場合
display(poke_DF['名前'])

# DataFrame型で抽出する場合
display(poke_DF[['名前']])
```

## 抽出条件で操作

```python
# ブールインデックス
poke_DF['タイプ1']=='くさ'

# ブールインデックスによる抽出
poke_DF[poke_DF['タイプ1']=='くさ']

poke_DF[(poke_DF['タイプ1']=='くさ') & (poke_DF['タイプ2']=='どく')]
```

## 統計量を調べる
- 平均
- 最大値
- 最小値
- 合計
- 標準偏差
- 分散
- 中央値
- 歪度
- 尖度

```python
max_HP = poke_DF['HP'].max()
poke_DF[poke_DF['HP'] == max_HP]
```

## 列単位で演算・加工

```python
poke_DF['合計種族値'] = poke_DF['HP'] + poke_DF['攻撃'] + poke_DF['防御'] + poke_DF['特攻'] + poke_DF['特防'] + poke_DF['素早さ']
display(poke_DF)
```

```python
poke_DF.groupby('名前').mean(numeric_only=True)
```
